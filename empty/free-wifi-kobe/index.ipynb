```python
import folium
import pandas as pd

df=pd.read_csv('opendata_94.csv',encoding='utf-8')

kobe_map = folium.Map(location=[34.728063, 135.23434], zoom_start=13)


for index, row in df.iterrows():
    folium.Marker(
        location=[row['緯度'], row['経度']],
        popup=f"{row['設置場所']} ({row['分類']})",
        icon=folium.Icon(icon="wifi", prefix='fa')
    ).add_to(kobe_map)


map_file_path = 'kobe_wifi_map.html'
kobe_map.save(map_file_path)


kobe_map
```
